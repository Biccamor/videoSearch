```python
model = AutoModel.from_pretrained("google/siglip2-base-patch16-224", dtype=torch.float16, device_map="cpu", attn_implementation="sdpa")
processor = AutoProcessor.from_pretrained("google/siglip2-base-patch16-224")
```

```python
image_dir = 'photos'
images = [img for img in os.listdir(image_dir)]
vector_images = []

for file_path in images:
    path = os.path.join(image_dir, file_path)
    try:
        image = Image.open(path)
        if image.mode != "RGB":
            image = image.convert("RGB")
        inputs = processor(images=image, return_tensors="pt")

        with torch.no_grad():
            image_features = model.get_image_features(**inputs)
            vector_images.append(image_features)
    except Exception as e: 
        print(f'błąd {e} w pliku {path}')
```
